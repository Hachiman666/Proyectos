```python
from google.colab import drive
import random
import tensorflow as tf
from tensorflow import random
import numpy as np
from keras import Sequential
from keras.layers import Flatten, Dense , Activation
from sklearn import preprocessing
from keras.utils import to_categorical
import pandas as pd
drive.mount("/content/drive/")
```

```python
df = pd.read_csv("/content/drive/MyDrive/iris.data", sep=",", header=None, names=["LS", "AS", "LP", "AP", "Clase"])

df = pd.concat([df, pd.get_dummies(df['Clase']).astype(int)], axis=1)

df.drop(columns=['Clase'], inplace=True)

X = df.iloc[:, :4].values
Y = df.iloc[:, 4:].values
```

```python
def ModeloEnt(X, Y, ncapas, nneuroxcapa, nepocas):
    model = Sequential()
    model.add(Dense(nneuroxcapa[0], activation='relu', input_dim=X.shape[1]))
    for i in range(1, ncapas):
        model.add(Dense(nneuroxcapa[i], activation='relu'))
    model.add(Dense(3, activation='softmax'))

    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    model.fit(X, Y, epochs=nepocas, batch_size=5)
    print("Entrenando...")
    loss, accuracy = model.evaluate(X, Y, verbose=0)
    print(f'Loss: {loss}, Accuracy: {accuracy}')

    Y_predict = model.predict(X)
    for i in range(len(X)):
        print(f'Registro : {i+1}, X: {X[i]}, Y: {Y[i]}, Y Predecida: {Y_predict[i]}')
```

```python
ncapas = int(input("Introduce el numero de capas: "))
nneuroxcapa = [int(input(f"Introduce el numero de las neuronas de las capas {i+1}: ")) for i in range(ncapas)]
nepocas = int(input("Introduce el numero de epocas que realizara la red neuronal: "))
ModeloEnt(X,Y,ncapas,nneuroxcapa,nepocas)
```
