```python
from google.colab import drive
import random
import tensorflow as tf
from tensorflow import random
import numpy as np
from keras import Sequential
from keras.layers import Flatten, Dense , Activation
from sklearn import preprocessing
from keras.utils import to_categorical
from keras.utils import plot_model
import pandas as pd
from sklearn.metrics import accuracy_score
from keras.callbacks import EarlyStopping
drive.mount("/content/drive/")
```

```python
df = pd.read_csv("/content/drive/MyDrive/glass.data", sep=",", header=None, names=["RI", "Na", "Mg", "Al", "Si", "K", "Ca", "Ba", "Fe", "Tipo"])
df = pd.concat([df, pd.get_dummies(df['Tipo']).astype(int)], axis=1)

df.drop(columns=['Tipo'], inplace=True)

X = df.iloc[:, :9].values
Y = df.iloc[:, 9:].values
```

```python
from sklearn.model_selection import train_test_split

def ModeloEnt(ncapas, nneuroxcapa, nepocas):
    Tabla = []
    for test_size in [i / 10.0 for i in range(1, 10)]:
      train_size = 1 - test_size
      print(f"Evaluando para test_size={test_size}...")
      X_train, X_test, y_train, y_test = train_test_split(X, Y,test_size=test_size, random_state=42)
      model = Sequential()
      model.add(Dense(nneuroxcapa[0], activation='relu', input_dim=X_train.shape[1]))
      for i in range(1, ncapas):
          model.add(Dense(nneuroxcapa[i], activation='relu'))
      model.add(Dense(6, activation='softmax'))

      model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
      print("Entrenando...")
      model.fit(X_train, y_train, epochs=nepocas, batch_size=32)
      print("Evaluando...")
      loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
      Tabla.append([train_size, test_size, loss, accuracy])


    Tabla.reverse()
    Tabla_df = pd.DataFrame(Tabla, columns=['Entrenamiento', 'Prueba', 'Loss', 'Accuracy'])
    print(Tabla_df)
```

```python
ncapas = int(input("Introduce el numero de capas: "))
nneuroxcapa = [int(input(f"Introduce el numero de las neuronas de las capas {i+1}: ")) for i in range(ncapas)]
nepocas = int(input("Introduce el numero de epocas que realizara la red neuronal: "))
ModeloEnt(ncapas,nneuroxcapa,nepocas)
print(f"Numero de capas: {ncapas}")
print(f"Numero de neuronas por capa: {nneuroxcapa}")
print(f"Numero de epocas: {nepocas}")
```
